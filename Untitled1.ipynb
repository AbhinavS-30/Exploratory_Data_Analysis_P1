```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == "USA")]
```

```python
df2.head()
```

```python
#cleaning athlete age
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#removing h from performance
```

```python
df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
 df2.head(10)
```

```python
#Removing some columns: Athlete club, Athlete country, Athlete year of birth, Athlete age category
```

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head(5)
```

```python
#cleaning null values
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#check for duplicate values
```

```python
df2[df2.duplicated() == True]
```

```python
#reset index
```

```python
df2.reset_index(drop = True)
```

```python
#fixing datatypes       
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2 = df2.rename(columns = {
'Year of event': 'year',
'Event dates': 'event_day',                   
'Event name' : 'event_name',                    
'Event distance/length' : 'distance',         
'Event number of finishers' : 'event_number_of_finishers',      
'Athlete performance' : 'athlete_performance',           
'Athlete gender' : 'athlete_gender',              
'Athlete average speed': 'athlete_average_speed',      
'Athlete ID' : 'athlete_id'                  
})
```

```python
df2.head(5)
```

```python
#reorder columns
```

```python
df3 = df2[['event_day', 'event_name', 'distance', 'event_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age','athlete_performance', 'athlete_average_speed']]
```

```python
#Lets answer some questions now
```

```python
#difference in speed for 50km, 50mi male to female
```

```python
df3.groupby(['distance', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
#Age groups best/worse in 50mi race (20 + races min)
```

```python
df3.query('distance == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19').head(15)
```

```python
df3.query('distance == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>9').head(15)
```

```python
#lets find seasons for runner. Summer slower? winter faster?
#winter = month 12-2
#spring = month 3-5
#summer = month 6-8
#fall = month 9-11
```

```python
df3['event_month'] = df3['event_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['event_season'] = df3['event_month'].apply(lambda x: 'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
```

```python
df3.head(25)
```

```python
df3.groupby('event_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
